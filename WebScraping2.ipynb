```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.arabam.com/ilan/galeriden-satilik-volkswagen-golf-1-6-tdi-trendline/firsat-araci-2012-model-volkswagen-golf-dizel-otomatik/24691302"

response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
containers = soup.find_all("div", class_="tab-content-car-information-container")


data = {}

for container in containers:

    title = container.find("h3").text.strip()


    if title in ["Genel Bakış", "Motor ve Performans", "Yakıt Tüketimi", "Boyut ve Kapasite"]:

        properties = container.find_all("li")


        property_data = {}

        # Her bir "property-key" ve "property-value" çifti için döngü
        for prop in properties:
            key = prop.find("span", class_="property-key").text.strip()
            value = prop.find("span", class_="property-value").text.strip()
            property_data[key] = value

        # Veri sözlüğüne bu özellikleri ve değerleri ekle
        data.update(property_data)

# Verileri DataFrame'e dönüştür
df_ozellik = pd.DataFrame(data, index=[0])

# "Özellik" başlığını ekleme ve alt başlıklarını bu başlık altına yerleştir
df_son = df_ozellik.rename(columns={"Genel Bakış": "Özellik"})
```
